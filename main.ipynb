### Walmart Sales Analysis:

```python
import pandas as pd
```

```python
df_walmart=pd.read_excel('WalmartSales.xlsx')
```

####   A. Analyze the performance of sales and revenue at the city and branch level 

```python
# Calculate total revenue for each transaction
df_walmart['Total_Revenue'] = df_walmart['Unit price'] * df_walmart['Quantity']
```

```python
Total_rev=df_walmart['Total_Revenue']
```

```python
# Group by City and Branch to calculate total sales and revenue
city_branch_performance = df_walmart.groupby(['City', 'Branch']).agg({
    'Total Revenue': ['sum', 'mean']
})
```

```python
city_branch_performance.columns = ['Total Sales', 'Average Sales']
```

```python
print(city_branch_performance)
```

#### B What is the average price of an item sold at each branch of the city 


```python
# Group by City and Branch to calculate average price
average_price_per_branch = df_walmart.groupby(['City', 'Branch']).agg({'Unit price': 'mean'})

print(average_price_per_branch)
```

```python
# Convert 'Date' column to datetime format
df_walmart['Date'] = pd.to_datetime(df_walmart['Date'])

# Extract month and year from the 'Date' column
df_walmart['Month'] = df_walmart['Date'].dt.month
df_walmart['Year'] = df_walmart['Date'].dt.year
```

```python
df_walmart['Date']
```

```python
# Group by Month, Product line, Gender, and Payment Method to calculate sales and revenue
monthly_performance = df_walmart.groupby(['Year', 'Month', 'Product line', 'Gender', 'Payment']).agg({
    'Total Revenue': 'sum',
    'Quantity': 'sum'
}).reset_index()
```

```python

monthly_performance
```

```python
# Filter data for April 2019
april_2019_data = monthly_performance[(monthly_performance['Year'] == 2019) | (monthly_performance['Month'] == 4)]
april_2019_data
```

```python
df_walmart.head(3)
```

```python
print(april_2019_data)
```

```python
# Identify focus areas for April 2019
focus_areas = april_2019_data.groupby(['Product line', 'Gender', 'Payment']).agg({
    'Unit price': 'sum',
    'Quantity': 'sum'
})
```
